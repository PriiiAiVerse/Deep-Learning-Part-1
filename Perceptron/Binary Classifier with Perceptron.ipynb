```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

weights = np.random.rand(2)
bias = np.random.rand()
lr = 0.1

for epoch in range(1000):
    for i in range(4):
        z = np.dot(X[i], weights) + bias
        pred = 1 if z > 0 else 0
        error = y[i] - pred
        weights += lr * error * X[i]
        bias += lr * error

print("Trained weights:", weights)
print("Trained bias:", bias)
```
